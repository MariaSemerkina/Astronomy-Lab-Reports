#### The Atlas with the absolute magnitudes in order to interpolate the datat for the specific classes

```python
specType=np.array([5,10,15,20,25,30,35,40,45,50,55,60,65,68,])
absomagatlas=np.array([-5.8,-4.1,-1.1,0.7,2,2.6,3.4,4.4,5.1,5.9,7.3,9,11.8,16])

```

```python
pars, cov= np.polyfit(specType,absomagatlas,3,cov=True)

def g(x):
    return pars[0]*x**3+pars[1]*x**2+pars[2]*x+pars[3]
#return pars[0]*x+pars[1]
    
#return pars[0]*x**3+pars[1]*x**2+pars[2]*x+pars[3]

xrange=np.arange(0,75)

plt.plot(xrange,g(xrange))
plt.scatter(specType,absomagatlas)
plt.title("Classification of Spectra of Stars")
plt.xlabel('Corresponding Spectral Classes')
plt.ylabel('Absolute Magnitude (M)')
plt.grid(True)
```

```python
#1-9 corresponds to the O class
#10-19 corresponds to the B class
#20-29 corresponds to the A class
#30-39 corresponds to the F class
#40-49 corresponds to the G class
#50-59 corresponds to the K class
#60-69 corresponds to the M class
```
