```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from scipy.optimize import curve_fit
%matplotlib inline
plt.rcParams.update({'font.size': 22})
plt.rcParams["figure.figsize"] = (15,10)
```

```python
def fitfunc(x,m,c):
    return m*x+c
```

#### The values for Distance, Velocity, Error of the velocity, and the Hubble constants to each galaxy.

```python
dis=np.array([211,72.4,565,295,594])
vel=np.array([15026.29,5990.68,39019.74,21024.40,41968.12])
velerr=np.array([564.63,510.53,457.4,510.53,610.11])

Hubblevals=np.array([71.21,82.74,69.06,71.27,70.65])
```

#### Calculating the mean of the Hubble constants in order to use for the Standard error formula.

```python
Hmean=np.mean(Hubblevals)
print(Hmean)
```

#### Code for calculting the standard error for the Hubble's constant value.

```python
i=0
for n in Hubblevals:
    i+=(n-Hmean)**2
s=np.sqrt(i/4)
SEH=s/np.sqrt(5)
print(SEH)
```

#### This code plots the graph below which includes error bars for velocity only, 
#### due to unmeasurbale uncertainties on the distance.

```python
plt.plot(dis,vel,'bo')
pars, cov = curve_fit(fitfunc,dis,vel)
plt.plot(dis,fitfunc(dis,*pars),'g-')

plt.title("Calculation of the Hubble Parameter, \nVelocity of Galaxies (Km/sec) vs Distance of the Galaxies (Mpc)")
plt.xlabel('Distance (Mpc)')
plt.ylabel('Velocity (km/sec)')
plt.grid(True)
plt.errorbar(dis,vel,velerr, fmt='o',capsize=5)
```

#### Using the covariance matrix to calculate the slope and error on the slope of the graph

```python
print('The slope is',pars[0],'km/sec/Mpc')
```

```python
print("the error on the slope is",np.sqrt(cov[0][0]),'')
```
