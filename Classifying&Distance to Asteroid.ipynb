```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.time import Time
from astropy.coordinates import SkyCoord, EarthLocation, AltAz
plt.rcParams.update({'font.size': 12})
plt.rcParams["figure.figsize"] = (10,8)
```

```python

astmove=SkyCoord(ra=['15:30:38.7','15:30:38.71','15:30:38.72','15:30:38.7','15:30:38.7','15:30:38.64','15:30:38.68'],
                 dec=['+11:14:06.2','+11:14:14.5','+11:14:18.3','+11:15:26.8','+11:15:34.6','+11:15:41.2','+11:15:54.7'],
              unit=(u.hourangle,u.deg))

star=SkyCoord(ra=['15:30:36.23','15:30:46.56','15:30:39.39'],dec=['+11:16:20.8','+11:15:0.8','+11:15:14.2'],
              unit=(u.hourangle,u.deg))


astpar=SkyCoord(ra=['15:30:37.72','15:30:38.69'],dec=['+11:15:36.1','+11:15:41.2'],unit=(u.hourangle,u.deg))

RAE=8e-5
DECE=8e-5

```

```python
plt.scatter(astmove.ra.deg,astmove.dec.deg,marker='+',color='red')
plt.scatter(star.ra.deg,star.dec.deg,marker='o',color='blue')
plt.errorbar(astmove.ra.deg,astmove.dec.deg,xerr=RAE,yerr=DECE,fmt='+',color='red')
plt.errorbar(star.ra.deg,star.dec.deg,xerr=RAE,yerr=DECE,fmt='+',color='blue')

plt.xlim((0.098+2.326e2),(0.05+2.326e2))
plt.xlabel('Right Ascension (\u00b0)')
plt.ylabel('Declination (\u00b0)')
plt.title('Movement of Asteroid')
plt.grid(True)

plt.text(star.ra.deg[0]-.0001,star.dec.deg[0],'RS1',horizontalalignment='left')
plt.text(star.ra.deg[1]-.0001,star.dec.deg[1],'RS2',horizontalalignment='left')
plt.text(star.ra.deg[2]-.0001,star.dec.deg[2],'RS3',horizontalalignment='left')

plt.text(astmove.ra.deg[0]-.0001,astmove.dec.deg[0],'B05',horizontalalignment='left')
plt.text(astmove.ra.deg[1]-.0001,astmove.dec.deg[1],'B07',horizontalalignment='left')
plt.text(astmove.ra.deg[2]-.0001,astmove.dec.deg[2],'B08',horizontalalignment='left')
plt.text(astmove.ra.deg[3]-.0001,astmove.dec.deg[3],'B09',horizontalalignment='left')
plt.text(astmove.ra.deg[4]-.0001,astmove.dec.deg[4],'B10',horizontalalignment='left')
plt.text(astmove.ra.deg[5]-.0001,astmove.dec.deg[5],'B12',horizontalalignment='left')
plt.text(astmove.ra.deg[6]-.0001,astmove.dec.deg[6],'B14',horizontalalignment='left')

```

```python
plt.scatter(astpar.ra.deg,astpar.dec.deg,marker='+',color='red')
plt.scatter(star.ra.deg,star.dec.deg,marker='o',color='blue')
plt.errorbar(star.ra.deg,star.dec.deg,xerr=RAE,yerr=DECE,fmt='+',color='blue')
plt.errorbar(astpar.ra.deg,astpar.dec.deg,xerr=RAE,yerr=DECE,fmt='+',color='red')


plt.xlim((0.1+2.326e2),(0.04+2.326e2))
plt.xlabel('Right Ascension (\u00b0)')
plt.ylabel('Declination (\u00b0)')
plt.title('Measuring the Parallax of the Asteroid')
plt.grid(True)

plt.text(astpar.ra.deg[0]-.0001,astpar.dec.deg[0],'ATEAST',horizontalalignment='left')
plt.text(astpar.ra.deg[1]-.0001,astpar.dec.deg[1],'ATWEST',horizontalalignment='left')

plt.text(star.ra.deg[0]-.0001,star.dec.deg[0],'RS1',horizontalalignment='left')
plt.text(star.ra.deg[1]-.0001,star.dec.deg[1],'RS1',horizontalalignment='left')
plt.text(star.ra.deg[2]-.0001,star.dec.deg[2],'RS1',horizontalalignment='left')
```
